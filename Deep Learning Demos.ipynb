## Short Tensorflow Tutorial

Before we run a simple neural network let's run a simple tensorflow demo to see how we might add two matrices together in a tensorflow session.

1. Initialize our Tensorflow and NumPy Libraries

2. Start the Tensorflow Session

3. Initialize our input variables ***input_1*** and ***input_2***

4. Define the math function to add inputs 1 and 2 and store in variable called ***output***

5. Evaluate output and store in result

6. Display result

```python
# Initialize our Tensorflow and NumPy libraries
import tensorflow as tf
import numpy as np

# Start Tensorflow session
with tf.Session():
  
  # Create inputs to model
  # Constant (used for biases) of value 1.0 in shape of 2x3 matrix
  input1 = tf.constant(1.0, shape=[2, 3])
  
  # Range of values between 1-6 in shape of 2x3 matrix
  input2 = tf.constant(np.reshape(np.arange(1.0, 7.0, dtype=np.float32), (2, 3)))
  
  # Add inputs 1 and 2 to output
  # Will add one to each value in the input2 matrix
  output = tf.add(input1, input2)
  
  # Evaluate and return result
  result = output.eval()

print(result)
```

## Tensorflow with MNIST

Now let's run a single simple forward-feed network to process images from the MNIST dataset (we'll come back to this dataset and run a demonstration of a more advanced network that will be able to make better predictions).

1. Load NumPy and Tensorflow libraries

2. Load the 'MNIST' digit image dataset

3. Define our Neural Network Weights

4. Define our Input and Output variables

5. Define our Error Reduction and Training Algorithms

6. Train Our Neural Network Model

7. Evaluate and Display the Performance of Model

```python
# Import Google's Tensorflow Deep Learning Library
import tensorflow as tf

# Load the MNIST imgage Dataset
mnist = tf.keras.datasets.mnist

# Split our data into training, testing, features, and labels
(x_train, y_train),(x_test, y_test) = mnist.load_data()

# Convert our input pixel features into values betweeon 0-1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define Single Layer Forward-feed Neural Network Model
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# Build our network and define our training methods and parameters
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the Model
model.fit(x_train, y_train, epochs=5)

# Evaluate and Display Accuracy
model.evaluate(x_test, y_test)
```
