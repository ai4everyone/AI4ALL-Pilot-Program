## Clustering

1. Load the sklearn and numpy libraries

2. Load (or define) our dataset

3. Define our unsupervised machine learning cluster model

4. Train (fit) the cluster model

5. Make predictions (will calculate the center of each cluster)

6. Visualize the dataset and the cluster centers

```python
# Use the sklearn library
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Load a dataset with two features, but no labels
X = np.array([[1, 2], [2, 1], [2, 0], [1, 1], [4, 2], [3, 3], [4, 1]])

# Choose a clustering model (will group data into two clusters)
kmeans = KMeans(n_clusters=2, random_state=0)

# Train the model
kmeans.fit(X)

# Return results of the model
kmeans.predict([[0, 0], [4, 4]])
print("The centers of our two clusters in our dataset would be {}:".format(kmeans.cluster_centers_))

# The x and y coordinates for the center of each cluster
x_centers = [3.66666667, 1.5]
y_centers = [2.0, 1.0]

# Visualize the centers of the data
fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(x = [1, 2, 2, 1, 4, 3, 4], y = [2, 1, 0, 1, 2, 3, 1], s = 150, linewidth = 4, color = 'black', marker = 'x');
ax.scatter(x_centers, y_centers)
```
